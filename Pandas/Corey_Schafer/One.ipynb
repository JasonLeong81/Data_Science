```python
# Video 6 
import pandas as pd 
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}

df = pd.DataFrame(people)

# combine first and last name 
df['full name'] = df['first'] + " " + df['last'] # we cannot use dot notation to make new data because python will think that we are adding a new attribute instead of adding a new data
df 

# removing first and last columns
df.drop(columns=['first','last'])

# do the same but reverse (split full name into first and last name)
df['full name'].str.split(' ',expand=True) # expand helps us split elements in list into columns # expands is only pandas not normal python 
df[['first','last']] = df['full name'].str.split(' ',expand=True)
df

# Add a single row 
df.append({'first':'Jason'},ignore_index=True) # others are set to NaN


# Combining data frame (multiple rows )
p = {
    "first": ["Jason", 'Chor', 'John'], 
    "last": ["Leong", 'Yew', 'Doe'], 
    "email": ["JasonLeong@gmail.com", 'choryew@email.com', 'JohnDoe@email.com']
}

df2 = pd.DataFrame(p)
df2
df = df.append(df2,ignore_index=True) # sort=False
df

# removing rows 
df.drop(index=4,inplace=True)
df

df.drop(index=df[df['last']=='Doe'].index)
df.drop(index=[1,2,5]) # same as previous line of code above 
```

```python
# Video 8
import pandas as pd 

df = pd.read_csv('survey_results_public.csv', index_col='Respondent')
schema_df = pd.read_csv('survey_results_schema.csv', index_col='Column')

df['ConvertedComp'].median() # median salary 
df.median() # only gives us back columns that are numbers 
df.describe() # same as in R 
df['ConvertedComp'].count() # how many rows are not empty (or NaN)
df['Hobbyist'].value_counts() # number of yes and number of no
schema_df.loc['SocialMedia'] 
df['SocialMedia'].value_counts()
df['SocialMedia'].value_counts(normalize=True) # percentage of peolpe using specific social media

# See specific results based on country 
df['Country'].value_counts() 
country_grp = df.groupby(['Country']) 
country_grp.get_group('India') 
filt = df['Country'] == 'India' 
df.loc[filt]['SocialMedia'].value_counts() # most popular social medias by country
country_grp['SocialMedia'].value_counts(normalize=True).loc['China'] # this returns a series with multiple indexs, thats why we can use .loc
country_grp['ConvertedComp'].median().loc['Germany'] # median salary in germnay # if loc only has one argument, thenit is row, otherwise we do :,'column name'
country_grp['ConvertedComp'].agg(['median', 'mean']) # use agg when you wanna use multiple statistical functions

# How many people in a country uses python
filt = df['Country'] == 'India' 
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum() # True as 1 and False as 0
# country_grp['LanguageWorkedWith'].str.contains('Python').sum() # error because?
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()) # number of people in each country who uses python

# What % of people from each country know python ? 
country_respondents = df['Country'].value_counts() # get total number of people who answered the survery
country_respondents 
country_uses_python = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum()) 
country_uses_python 
python_df = pd.concat([country_respondents, country_uses_python], axis='columns', sort=False) # concat can combine more than one series together # axis = column so that series don't get combined as rows
python_df 
python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True) #
python_df 
python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100 # doing the division to get the percentage
python_df 
python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)
python_df.head(50) 
python_df.loc['Japan'] 

```

```python
# Video 9 
import pandas as pd 
import numpy as np 
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

df = pd.DataFrame(people)

df.dropna(axis='index',how='any') # index AND any means srop rows which have at least one na # index AND all drops rows that have all values na # axis can take index or columns and how can take any or all
df.dropna(axis='index',how='all',subset=['last','email']) # for a row to be dropped, both of the subsets last and email has to be empty # can use inplace here

df = pd.DataFrame(people)
df
df.replace('NA',np.nan,inplace=True)
df.replace('Missing',np.nan,inplace=True) # all 'Missing' becomes NaN
df

df.isna() # return us a True/False
df.fillna(0) # change all na to 0

### check and solve for error
df.dtypes # not a method, its an attribute
# df['age'].mean() # won't work because age is in string
# Nan is actually a float so we have to use float when dealing with NaN
# df['age'] = df['age'].astype(int) # we get an error
# df = pd.DataFrame(people)
# df['age'] = df['age'].astype(float) # why???
# df.dtypes

na_vals = ['NA', 'Missing']
df = pd.read_csv('survey_results_public.csv', index_col='Respondent', na_values=na_vals)
schema_df = pd.read_csv('survey_results_schema.csv', index_col='Column')

df['YearsCode'].head(5)
# df['YearsCode'] = df['YearsCode'].astype(float) # 'less than 1 year'
df['YearsCode'].unique() # see all unique values
df['YearsCode'].replace('Less than 1 year', 0, inplace=True)
df['YearsCode'].replace('More than 50 years', 51, inplace=True)
df['YearsCode'] = df['YearsCode'].astype(float) # now it works 
df['YearsCode'].mean()
```
